```python
import math
# times = probe_size/2048 / chunk_queue_thr
# each time data_amount = sel * probe_size /times
#input :  (data_amount, build_size, sel * probe_size)
#values: build_size, payload_size, sel, chunk_queue_thr
probe_size = 20000000
def chunk_num(values):
    sel = values[2]
    if sel ==1:
        return 100
    else:
        return (probe_size/2048)
def map_build(values):
    sel = values[2]
    times = chunk_num(values)/values[-1]
    data_amount = sel * probe_size /times
    build_size = values[0]
    total_amount = sel * probe_size
    return max(times * (0.2315086510355684 + 7.05514525e-05 * data_amount + 
                    2.93549977e-02 * build_size/122880 - 3.20315483e-06 * total_amount ),0)
def materialize_time(values):
    sel = values[2]
    times = chunk_num(values)/values[-1]
    data_amount = sel * probe_size /times
    payload = values[1]
    return max(times * (20.6899376233236 + 1.24389500e-04* data_amount + 
                    1.50300152*payload + 8.21514162e-07 * (payload * data_amount) + 2.40712219e-05 * math.log2(payload*122880/(256*1024)) ),0)
def hash_build(values,pay=0):
    build_size = values[0]
    payload = values[1]
    if pay!=0:
        payload = pay
    return max(5.11123454e-05 * build_size + 2.24206187 * payload - 85.1251751093121,0)
def cal_em_pred(values):
    return hash_build(values)
def cal_lm_pred(values):
    # print(hash_build(values,8),map_build(values),materialize_time(values))
    total = hash_build(values,8)+map_build(values)+materialize_time(values)
    return total
```

```python
import numpy as np
import pandas as pd
from pandas.core.frame import DataFrame
training_data = []
lm = []
ff = open('/home/yihao/duckdb/ht/duckdb/examples/embedded-c++/release/payload_build_time.txt', 'r')
f = [l[:-1] for l in ff.readlines()]
correct =0
total =0
for i in range(0,len(f),11):
    line = f[i]
    values = list(map(float, line.split()))
    em_time = float(f[i+1].split()[-1])
    idx=2
    for chunk_thr in [1,10,50,100,500,1000,2000,5000,10000]:
        lm_time = float(f[i+idx].split()[-1])
        choose_lm = lm_time<em_time
        total+=1
        idx+=1
        value_tmp = list(map(float, line.split()))
        value_tmp.append(chunk_thr)
        em = round(cal_em_pred(value_tmp),2)
        lm = round(cal_lm_pred(value_tmp),2)
        train= list(map(float, line.split()))
        train.append(chunk_thr)
        train.append(choose_lm)
        training_data.append(train)
        
train = DataFrame(training_data)
train
```

```python

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

X = train.iloc[:,:4]
y = train.iloc[:,4]
print(X)
# 拆分数据集为训练集和测试集
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

# 标准化特征
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)

# 选择分类器并训练模型
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)

# 预测测试集
y_pred = classifier.predict(X_test)

# 评估模型性能
accuracy = accuracy_score(y_test, y_pred)
conf_matrix = confusion_matrix(y_test, y_pred)
class_report = classification_report(y_test, y_pred)

print(f"Accuracy: {accuracy}")
print("Confusion Matrix:")
print(conf_matrix)
print("Classification Report:")
print(class_report)
```
